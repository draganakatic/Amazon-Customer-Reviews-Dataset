```python
import numpy as np
import pandas as pd

```

```python
data=pd.read_csv(r'C:\Users\draga\OneDrive\Documents\Desktop\Reviews.csv')
data.describe()
```

```python
data.head()
```

```python
data = data[data['Score'] != 3]

# napravi novu kolonu Sentiment
data['Sentiment'] = data['Score'].apply(lambda x: 'positive' if x > 3 else 'negative')


```

```python
print(data['Sentiment'].value_counts())
data.head()
```

```python
sample_size = 10000
half_size = sample_size // 2

negative = data[data['Sentiment'] == 'negative'].sample(n=half_size, random_state=42)
positive = data[data['Sentiment'] == 'positive'].sample(n=half_size, random_state=42)

data = pd.concat([negative, positive]).sample(frac=1, random_state=42)

print(data['Sentiment'].value_counts())
print(data.shape)
data.head()
```

```python
# saƒçuvaj kao CSV fajl
data.to_csv('Reviews_sample_balanced.csv', index=False)

print("Fajl sacuvan kao Reviews_sample_balanced.csv")
```
